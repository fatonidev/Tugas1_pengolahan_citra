```python
# pip install numpy imageio matplotlib
import numpy as np
import imageio as img
import matplotlib.pyplot as plt

# Membaca gambar
image = img.imread("/content/pngtree-the-colored-gas-filled-balloons-attached-to-the-yarn-are-flying-image_15917591.jpg")

# Memisahkan channel warna
red = image[:, :, 0]
green = image[:, :, 1]
blue = image[:, :, 2]

# Membuat gambar untuk setiap channel
imgRed = np.zeros_like(image)
imgGreen = np.zeros_like(image)
imgBlue = np.zeros_like(image)

imgRed[:, :, 0] = red       # Channel Red
imgGreen[:, :, 1] = green   # Channel Green
imgBlue[:, :, 2] = blue     # Channel Blue

# Menampilkan channel Green dan Blue
plt.figure(figsize=(10, 10))

# Channel Green
plt.subplot(3, 1, 1)
plt.imshow(imgGreen)
plt.title("Green Channel")

# Channel Blue
plt.subplot(3, 1, 2)
plt.imshow(imgBlue)
plt.title("Blue Channel")

plt.show()
```
